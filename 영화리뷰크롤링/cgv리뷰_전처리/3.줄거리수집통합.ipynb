```python
import pandas as pd
import os
import glob
```

# 1. 통합

```python
origin_folder="../영화리뷰데이터_저장/cgv_장르_줄거리/"
search_pattern = os.path.join(origin_folder, '*.csv')
file_list=glob.glob(search_pattern)
```

```python
result=pd.DataFrame()
j=0
for i in file_list:
    j+=1
    temp=pd.read_csv(i, encoding='utf-8-sig')
    result=pd.concat([result,temp])
    print(f"{j}/{len(file_list)} complete : {i}")
```

# 2. 전처리

```python
result.isna().sum()
```

```python
result.info()
```

```python
result.isna().value_counts().reset_index()
```

줄거리만 결측치를 제거하자

```python
result.dropna(subset=['줄거리'], inplace=True)
```

```python
result.head(2)
```

```python
# 특수문자 제거
result.loc[:,'줄거리'] = result['줄거리'].str.replace('\n', ' ').str.replace('\r', ' ').str.replace('\t', ' ')
result.head(2)
```

```python
result.to_csv('전처리파일저장/cgv_장르_줄거리_통합본.csv', index=False, encoding='utf-8-sig')
```
