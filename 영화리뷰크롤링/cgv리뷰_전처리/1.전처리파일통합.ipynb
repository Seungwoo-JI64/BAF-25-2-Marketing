```python
import pandas as pd
import os
import glob
import tqdm
```

# 1. 개별 파일 통합

```python
origin_folder="../영화리뷰데이터_저장/cgv데이터/"
search_pattern = os.path.join(origin_folder, '*.csv')
file_list=glob.glob(search_pattern)
```

```python
file_list.sort()
```

```python
result=pd.DataFrame()
j=0
for i in tqdm.tqdm(file_list):
    j+=1
    temp=pd.read_csv(i, encoding='utf-8-sig')
    result=pd.concat([result,temp])
    print(f"{j}/{len(file_list)} complete : {i}")
```

```python
result.head()
```

```python
result.to_csv("전처리파일저장/cgv_원본_통합파일.csv", index=False, encoding='utf-8-sig')
```
