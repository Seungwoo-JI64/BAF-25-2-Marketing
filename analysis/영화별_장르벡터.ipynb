```python
import pandas as pd
```

```python
movie_info = pd.read_csv("movie_info_data.csv", encoding="utf-8-sig")
```

```python
movie_info = movie_info.iloc[:, [0,-7,-6,-5,-4,-3,-2,-1]]
movie_info.head()
```

```python
# 멀티핫인코딩을 위해 모든 장르 컬럼을 하나로 합친 후, 장르별로 멀티핫 벡터 생성
genres = movie_info[['genre1', 'genre2', 'genre3', 'genre4', 'genre5', 'genre6', 'genre7']]
genre_set = pd.unique(genres.values.ravel('K'))
genre_set = [g for g in genre_set if pd.notnull(g)]
# 각 영화의 장르를 멀티핫 벡터로 변환
movie_info['genre_multi_hot'] = genres.apply(
    lambda row: [int(g in row.values) for g in genre_set], axis=1
)
```

```python
movie_info[["Movie_Title", "genre_multi_hot"]].to_csv("movie_genre_multi_hot.csv", index=False, encoding="utf-8-sig")
```
