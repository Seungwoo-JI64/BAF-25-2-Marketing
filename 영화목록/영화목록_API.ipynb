## 1. API 전체 목록 불러오기

```python
# 여기에 발급받은 API 키를 입력하세요.
API_KEY = "0fedb3690ed2dbeb688fb9f77f9b7fdc"

# 영화 목록 API의 기본 URL
BASE_URL = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json"
```

```python
# 전체 영화 데이터를 저장할 리스트 초기화
all_movies_list = []

# 페이지 처리를 위한 변수 초기화
cur_page = 1
item_per_page = 100 # 한 페이지에 표시할 항목 수 (최대값으로 설정 권장)

print("영화 목록 데이터 추출을 시작합니다...")

while True:
    # API 요청에 필요한 파라미터 설정
    params = {
        'key': API_KEY,
        'curPage': str(cur_page),
        'itemPerPage': str(item_per_page),
    }

    try:
        # API 요청 보내기
        response = requests.get(BASE_URL, params=params)
        response.raise_for_status() # 오류가 발생하면 예외를 발생시킴

        # JSON 응답 파싱
        data = response.json()
        movie_list = data.get('movieListResult', {}).get('movieList', [])

        # 더 이상 가져올 영화 목록이 없으면 반복 종료
        if not movie_list:
            print("더 이상 가져올 데이터가 없어 추출을 종료합니다.")
            break

        # 현재 페이지의 영화 목록을 전체 리스트에 추가
        all_movies_list.extend(movie_list)

        print(f"페이지 {cur_page} 처리 완료 | 수집된 영화 수: {len(all_movies_list)}개")

        # 다음 페이지로 이동
        cur_page += 1

        # 서버 부하를 줄이기 위한 짧은 대기 시간
        time.sleep(0.5)

    except requests.exceptions.RequestException as e:
        print(f"API 요청 중 오류 발생: {e}")
        break
    except Exception as e:
        print(f"알 수 없는 오류 발생: {e}")
        break

print(f"\n총 {len(all_movies_list)}개의 영화 데이터를 성공적으로 수집했습니다.")
```

```python
movie_data=pd.DataFrame(all_movies_list)
```

# 박스오피스 상위 10개 정보

```python
BASE_URL1 = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json"

# 💡 조회하고 싶은 날짜를 'YYYYMMDD' 형식으로 바꾸면 됩니다.
# 여기서는 어제 날짜인 2025년 9월 6일로 설정했습니다.
target_dt = "20250906" 

# 3. API 요청 보내기
# 요청에 필요한 파라미터들을 딕셔너리 형태로 구성합니다.
params = {
    'key': API_KEY,
    'targetDt': target_dt,
    'itemPerPage': '100'  # 상위 10개 영화만 조회
}

try:
    # requests.get() 함수를 사용해 API에 GET 요청을 보냅니다.
    response = requests.get(BASE_URL1, params=params)
    response.raise_for_status() # HTTP 오류가 발생하면 예외를 일으킵니다.

    # 4. 응답 데이터 처리 (JSON -> Pandas DataFrame)
    # 응답받은 JSON 데이터를 파이썬 딕셔너리로 변환합니다.
    data = response.json()

    # 실제 박스오피스 목록 데이터만 추출합니다.
    movie_list = data['boxOfficeResult']['dailyBoxOfficeList']

    # 보기 편하도록 딕셔너리 리스트를 pandas DataFrame(표)으로 변환합니다.
    df = pd.DataFrame(movie_list)
    
    # 5. 결과 출력
    print(f"✅ {target_dt} 일별 박스오피스 TOP 10\n")
    
    # 필요한 컬럼만 선택하여 순서대로 정렬 후 출력할 수 있습니다.
    # 예: 순위, 영화명, 개봉일, 당일 관객수, 누적 관객수
    columns_to_show = ['rank', 'movieNm', 'openDt', 'audiCnt', 'audiAcc']
    display(df[columns_to_show])

except requests.exceptions.RequestException as e:
    print(f"❌ API 요청 중 오류가 발생했습니다: {e}")
except KeyError:
    # API 응답 구조가 예상과 다를 때를 대비한 예외 처리
    print("❌ 응답 데이터의 구조가 예상과 다릅니다. 'boxOfficeResult' 또는 'dailyBoxOfficeList' 키를 찾을 수 없습니다.")
except Exception as e:
    print(f"❌ 알 수 없는 오류가 발생했습니다: {e}")
```
