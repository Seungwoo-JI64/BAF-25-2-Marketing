```python
# -----------------------------
# 유틸 함수
# -----------------------------
def _parse_titles(s, sep="|"):
    """'제목1|제목2|...' 문자열을 ['제목1','제목2', ...]로 변환"""
    if pd.isna(s):
        return []
    return [t.strip() for t in str(s).split(sep) if t.strip()]

def _ensure_list_of_ints(x):
    """
    genre_multi_hot이 object형(리스트) 또는 문자열('[0,1,0,...]')이어도
    안전하게 정수 리스트로 변환
    """
    if x is None or (isinstance(x, float) and np.isnan(x)):
        return None
    if isinstance(x, list):
        return [int(v) for v in x]
    if isinstance(x, np.ndarray):
        return [int(v) for v in x.tolist()]
    if isinstance(x, str):
        s = x.strip()
        try:
            arr = ast.literal_eval(s)  # '[0,1,0]' → [0,1,0]
            return [int(v) for v in arr]
        except Exception:
            # '0,1,0' 같은 형태 대비
            s = s.strip("[]")
            parts = [p.strip() for p in s.split(",") if p.strip() != ""]
            return [int(float(p)) for p in parts]
    # 그 외 타입은 실패 처리
    raise ValueError(f"genre_multi_hot 값을 리스트로 변환할 수 없습니다: {x}")

def _l1_normalize(vec: np.ndarray) -> np.ndarray:
    """합이 1이 되도록 L1 정규화(합이 0이면 그대로 반환)"""
    s = vec.sum()
    return vec / s if s > 0 else vec

# -----------------------------
# 메인 파이프라인
# -----------------------------
def build_actor_and_movie_vectors(actor_df: pd.DataFrame, genre_df: pd.DataFrame):
    """
    1) 배우별 filmography를 genre_df와 매칭 → 장르 멀티핫 합산→ L1 정규화 → actor_genre_vector
    2) 각 영화 기준 출연 배우 구하기 (movie_to_actors)
    3) 영화별 출연 배우의 actor_genre_vector 평균(=합산 후 L1 정규화) → movie_actor_genre_vector
    """
    # --- genre_df 전처리: 장르 벡터 확정 & 길이 점검
    gdf = genre_df.copy()
    gdf["__vec"] = gdf["genre_multi_hot"].apply(_ensure_list_of_ints)
    # 벡터 길이(장르 차원) 확인
    vec_lens = gdf["__vec"].dropna().map(len).unique()
    if len(vec_lens) != 1:
        raise ValueError(f"genre_multi_hot 벡터 길이가 일관되지 않습니다: {vec_lens}")
    dim = int(vec_lens[0])
    # 제목→벡터 매핑(dict)
    title_to_vec = {
        str(t).strip(): np.array(v, dtype=float)
        for t, v in zip(gdf["Movie_Title"], gdf["__vec"])
    }

    # --- actor_df 전처리: 배우별 필모그래피를 리스트로 만들고, 중복 영화 제거
    adf = actor_df.copy()
    adf["__titles"] = adf["filmoNames"].apply(_parse_titles)
    # 같은 배우 이름이 여러 행에 있을 수 있으니 통합(유니온)
    actor_to_titles = (
        adf.groupby("peopleNm")["__titles"]
           .agg(lambda lists: sorted({t for lst in lists for t in lst}))
           .to_dict()
    )

    # --- 1) 배우별 actor_genre_vector 계산
    actor_rows = []
    movie_to_actors = defaultdict(list)

    for actor, titles in tqdm(actor_to_titles.items()):
        # genre_df에 존재하는 영화만 매칭
        matched = [t for t in titles if t in title_to_vec]
        vecs = [title_to_vec[t] for t in matched]

        if len(vecs) > 0:
            sum_vec = np.sum(vecs, axis=0)           # 예: [2,0,1,0,1]
            actor_vec = _l1_normalize(sum_vec)       # 예: [0.5,0,0.25,0,0.25]
        else:
            # 매칭 영화가 없다면 NaN 벡터로 표기(필요 시 0 벡터로 바꿔도 됨)
            actor_vec = np.full(dim, np.nan)

        actor_rows.append({
            "peopleNm": actor,
            "matched_titles": matched,  # 이 배우가 genre_df에서 매칭된 영화 목록
            "num_matched": len(matched),
            "actor_genre_vector": actor_vec.tolist(),
        })

        # 영화→배우 역매핑
        for t in matched:
            movie_to_actors[t].append(actor)

    actor_genre_df = pd.DataFrame(actor_rows)

    # --- 2) (이미 위에서) movie_to_actors 완성

    # --- 3) 영화별로 출연 배우들의 actor_genre_vector 평균(=합산 후 L1 정규화)
    movie_rows = []
    # genre_df 기준의 모든 영화에 대해 계산
    for title in tqdm(gdf["Movie_Title"].astype(str).map(str.strip)):
        actors = movie_to_actors.get(title, [])
        vecs = []
        for a in actors:
            row = actor_genre_df.loc[actor_genre_df["peopleNm"] == a, "actor_genre_vector"]
            if not row.empty:
                v = np.array(row.iloc[0], dtype=float)
                if not np.isnan(v).all():
                    vecs.append(v)

        if len(vecs) > 0:
            sum_vec = np.sum(vecs, axis=0)
            movie_actor_vec = _l1_normalize(sum_vec)  # 배우 기여 벡터들의 합 → L1 정규화
        else:
            movie_actor_vec = np.full(dim, np.nan)

        movie_rows.append({
            "Movie_Title": title,
            "cast_peopleNm": actors,
            "n_cast": len(actors),
            "movie_actor_genre_vector": movie_actor_vec.tolist(),
        })

    movie_actor_genre_df = pd.DataFrame(movie_rows)

    return actor_genre_df, movie_actor_genre_df, dict(movie_to_actors)
```

```python
actor_genre_df, movie_actor_genre_df, movie_to_actors = build_actor_and_movie_vectors(actor_df, genre_df)
```
